```python
# 감마 보정
# 브라운관의 단점 : 비선형적 
# 인간의 시각 : 비선형적
# 감마값을 조정하여 이미지의 밝기를 조절하는 작업
# 감마값 1보다 크면 밝아지고, 작으면 어두워짐

import cv2
import matplotlib.pyplot as plt

img = cv2.imread("c:/data/images/peach.png")
cv2.imshow("original", img)  # 타이틀, 이미지
cv2.waitKey(0) # 아무키나 입력하면 창이 닫힘
cv2.destroyAllWindows()
```

```python
image_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# 이미지 화면에 표시
plt.imshow(image_rgb)
#plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
import numpy as np

plt.figure(figsize=(20,5))

gamma = np.arange(0.1, 2.0, 0.2)

for i, g in enumerate(gamma):
    plt.subplot(2, 5, i+1)
    out = image_rgb.copy() # 이미지 복사
    out = image_rgb.astype(np.float64)
    out = ((out / 255) ** (1 / g)) * 255 # 이미지 보정 공식
    out = out.astype(np.uint8)
    plt.title(f'gamma:{g:.1f}')
    plt.imshow(out)
    plt.axis("off")
plt.show()
```
